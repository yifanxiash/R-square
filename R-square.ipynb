##R-square demostrate all the details on how to calculate R-sqaure in a linear regression model
##y = b1 * x + b0

###Residual sum of squares(RSS): RSS = (y1 - b0 - b1 * x1)^2 + ......+ (yn - b0 -b1 * xn )^2
###Residual Standard Error: RSE = ((RSS/(n-2))) ^ (1/2)
###Total sum of quares TSS = (y1 - y)^2 + ...... + (yn - y)^2
###R-square = (TSS - RSS) / TSS = 1 - RSS/TSS

##R-square measures the propoition of variability in Y that can be exmplanined using X
###An R-square that is close to 1 indicates that a large proportion of the variablity in the response has been 
###explanined by the regression, and a number near 0 indicates that the regression did not expalnin muvh of 
###the variability in the response

##R-square is a measure of the model's quality and the Adjusted R-square account fot the number of variable in
##the model and is a more realistic assessment of its effectiveness, which introduces a penalty for the number
##of estimated cifficients 
###adjusted R-square = 1 - (RSS/(n - p -1)) / (tss/(n - 1))


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn
seaborn.set()
import numpy as np
import sklearn
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Generate random testing sample
n = 100
x = 2 * np.random.rand(n,1)
y = 4 + 3.5 * x + np.random.randn(n,1)
plt.figure(figsize = (15,8))
plt.scatter(x,y)
```

```python
#Fitting the dataset with linearregression model in scikit learn
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(x,y) 
print("b0 = " + str(float((lin_reg.intercept_)))[:5] + "  b1 = " + str(float(lin_reg.coef_))[:5])
y_predict = lin_reg.predict(x)
y_predict
plt.figure(figsize = (15,8))
plt.scatter(x,y)
plt.plot(x,y_predict)
plt.show()
```

```python
##Using sklearn built-in function to find value of those metrics
print("R-square:" + str(sklearn.metrics.r2_score(y, y_predict))[:7])
#sklearn.metrics.rse_score(y, y_predict)
```

```python
##Calculatin R-square step by step
RSS = np.sum(np.square(np.array(y - y_predict)))
print("RSS: " + str(RSS)[:7])
RSE =np.sqrt( RSS / (n - 2))
print("RSE: " + str(RSE)[:7])
#TSS = np.sum()
TSS = np.sum(np.square(np.array(y - np.average(y))))
print("TSS: " + str(TSS)[:7])
R_square = 1 - RSS/TSS #which is telling us 81.67% of variability in Y that can be exmplanined using X
print("R_square: " + str(R_square)[:7]) 
adj_R_square = 1 - (RSS/(n - 1 - 1))/(TSS/(n - 1))
print("adj_R_square: " + str(adj_R_square)[:7])
```
